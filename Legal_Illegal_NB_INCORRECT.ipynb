```python
import pandas as pd
import numpy as np
import time  
from sklearn import metrics  
import pickle as pickle  
from sklearn.cross_validation import train_test_split 
```

```python
def read_data(data_file_path):  
    
    training_df = pd.read_csv(data_file_path, header=0)
    feature_cols = ['T0101','T0102','T0201','T0202','T0203','T0204','T0205','T0206','T0207','T0208','T0209','T0210','T0211','T0212','T0213',
'T0301','T0302','T0303','T0304','T0305','T0306','T0307','T0308','T0309']
    x = training_df[feature_cols]
    y = training_df['LEGAL_RECORD']
    x_train,x_test, y_train, y_test = train_test_split(x, y, random_state=1) 
    
    return training_df, x_train, y_train, x_test, y_test
```

```python
data_file = "C:/Users/chenyuxiang/Desktop/Legal_Illegal_Record_all_data.csv"  

all_data_df, train_x, train_y, test_x, test_y = read_data(data_file)
```

```python
# Multinomial Naive Bayes Classifier  
def naive_bayes_classifier(train_x, train_y):  
    from sklearn.naive_bayes import MultinomialNB  
    model = MultinomialNB(alpha=0.01)  
    model.fit(train_x, train_y)  
    return model 
```

```python
NB_model = naive_bayes_classifier(train_x, train_y)
```

```python
predict_y_array = NB_model.predict(test_x)
```

```python
predict_y = pd.DataFrame({'Legal_Predict': predict_y_array})
predict_y.index = test_y.index

test_y_df = test_y.to_frame()
```

```python
result_y_df = pd.merge(predict_y, test_y_df, left_index=True, right_index=True)
```

```python
result_df = pd.merge(result_y_df, test_x, left_index=True, right_index=True)
```

```python
predict_incorrect_df = result_df[result_df['Legal_Predict'] != result_df['LEGAL_RECORD']]
```

```python
all_data_result_df = pd.merge(predict_y, all_data_df, left_index = True, right_index = True)
```

```python
all_data_predict_incorrect_df = all_data_result_df[all_data_result_df['Legal_Predict'] != all_data_result_df['LEGAL_RECORD']]
```
